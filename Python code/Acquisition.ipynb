# Data acquisition

First of all, we get the data from Twitter, through scraping. The tweets we're looking for contain several hashtag regarding the Qatar World Cup 2022. We select only those tweets with 5+ likes.

```python
# Creating list to append tweet data to
tweets_list = []

# Performing the query
query = """#Qatar2022 OR #WorldCup2022 OR #BoycottQatar2022 OR #BoycottQatar OR 
#FIFAWorldCup2022 OR #World2022 OR #QatarWorldCup2022 OR #BoycottWorldCup2022 
OR #NotMyWorldCup OR #WC2022 OR #BoycottQatarWorldCup2022 OR #BoycottWorldCup 
OR #BoycottQatarWC OR #BoycottQatar OR #PulseQatar22 OR #PulseWorldCup2022 OR 
#WCQ2022 OR #WorldCupQatar2022 OR #QatarWorldCup OR (#Qatar (#WorldCup OR #FIFA 
OR #WorldCupDraw OR #QatarGate OR #WorldCupQualifiers OR #WCQ)) 
until:2022-12-19 lang:en min_faves:5"""
       
# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
    if(i%10000 == 0):
        print(i)
    tweets_list.append([tweet.date, tweet.content, tweet.user.username, tweet.likeCount, tweet.retweetCount])
    
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns = ['Datetime', 'Text', 'Username', 'Likes', 'Retweets'])

# Saving list of tweets to csv
tweets_df.to_csv('tweets.csv', index = False)
```

Now, let's count the most popular hashtag trends since the announcement of the Qatar World Cup.

```python
df = pd.read_csv('gdrive/My Drive/Text Mining & Search/tweets.csv')
df['Hashtags'] = df['Text'].apply(lambda x: re.findall(r"#(\w+)", x))
# Remove duplicates
df['Hashtags'] = df['Hashtags'].apply(lambda x: list(dict.fromkeys(x)))
```

```python
hashtag_lists = df['Hashtags'].tolist()
```

```python
hashtag_list = hashtag_lists[0]
for i in range(1, len(hashtag_lists)):
  hashtag_list+=hashtag_lists[i]
```

```python
hashtag_list = Counter(hashtag_list).most_common()
```

```python
hash_df = pd.DataFrame(hashtag_list, columns = ['Hashtag', 'Occurrences'])
```
